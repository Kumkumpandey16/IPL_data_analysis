```python
df["date"]=pd.to_datetime(df["date"])
```

```python
df["date"].min()
```

```python
df["venue"].nunique()
```

```python
df["venue"].unique()
```

```python
df["venue"].value_counts().head(10).plot(kind="bar")
plt.title("Top 10 Venues")
plt.xlabel("Venue")
plt.ylabel("Matches")
plt.show()
```

The majority of IPL matches have been played at a small number of venues, indicating that certain stadiums host games more frequently.
